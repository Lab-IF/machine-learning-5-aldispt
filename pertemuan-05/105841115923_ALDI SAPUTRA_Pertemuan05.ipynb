# Pertemuan 5: Support Vector Machine (SVM)

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. Load & Split Dataset
# ============================================================
# Menggunakan Breast Cancer Wisconsin dataset (Klasifikasi Biner: Malignant/Benign)
data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)

#

# Scaling (Wajib untuk SVM)
# SVM mencari margin maksimum antar kelas. Jika skala fitur berbeda (misal: 0-1 vs 0-1000),
# fitur dengan skala besar akan mendominasi perhitungan jarak.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Tugas 1: Bandingkan Kernel Linear vs RBF
# ============================================================
# Kernel trick memungkinkan SVM bekerja pada ruang dimensi tinggi tanpa menghitung koordinat eksplisit.
models = {
    "SVM Linear": SVC(kernel="linear", C=1),
    "SVM RBF": SVC(kernel="rbf", C=1, gamma="scale"),
}

print("=" * 60)
print("Tugas 1: Perbandingan Kernel Linear vs RBF")
print("=" * 60)

#

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    acc = round(accuracy_score(y_test, pred), 3)
    results[name] = acc
    print(f"\n{name} Accuracy: {acc}")
    print(f"Confusion Matrix {name}:")
    print(confusion_matrix(y_test, pred))
    print(f"Classification Report {name}:")
    print(classification_report(y_test, pred, target_names=data.target_names))
```

```python
# Tugas 2: Tuning Sederhana (C dan Gamma)
# ============================================================
# C: Kontrol trade-off antara margin lebar dan klasifikasi yang benar (Regularisasi).
# Gamma: Menentukan seberapa jauh pengaruh satu contoh pelatihan mencapai (Hanya untuk RBF/Poly).
print("\n" + "=" * 60)
print("Tugas 2: Parameter Tuning Analysis (C dan Gamma)")
print("=" * 60)

#

best_acc = 0
best_params = {}

# Mencari kombinasi terbaik secara manual (mirip Grid Search sederhana)
for c_val in [0.1, 1, 10]:
    for gamma_val in [0.01, 0.1, 1]:
        model = SVC(kernel="rbf", C=c_val, gamma=gamma_val)
        model.fit(X_train, y_train)
        y_pred_tune = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred_tune)

        print(f"  C={c_val:>4}, gamma={gamma_val:>4} → Accuracy: {acc:.3f}")

        if acc > best_acc:
            best_acc = acc
            best_params = {"C": c_val, "gamma": gamma_val}

print(f"\n→ Parameter Terbaik Hasil Tuning: C={best_params['C']}, gamma={best_params['gamma']}")
print(f"→ Akurasi Tertinggi: {best_acc:.3f}")
```

```python
# Tugas 3: Kesimpulan & Interpretasi
# ============================================================
print("\n" + "=" * 60)
print("Tugas 3: Kesimpulan")
print("=" * 60)
print("1. Pentingnya Scaling: Tanpa StandardScaler, performa SVM akan menurun drastis")
print("   karena algoritma ini berbasis perhitungan jarak (Euclidean).")
print(f"2. Efektivitas Kernel: {'SVM RBF' if results['SVM RBF'] >= results['SVM Linear'] else 'SVM Linear'}")
print("   cenderung memberikan hasil lebih baik karena mampu menangkap pola non-linear.")
print("3. Pengaruh Parameter:")
print(f"   - Nilai C yang terlalu besar dapat menyebabkan overfitting (margin sempit).")
print(f"   - Nilai Gamma yang terlalu tinggi membuat model terlalu sensitif pada data point terdekat.")
print(f"4. Rekomendasi: Gunakan C={best_params['C']} dan Gamma={best_params['gamma']} untuk model produksi.")

print("\n✅ Pertemuan 05 selesai.")
```
