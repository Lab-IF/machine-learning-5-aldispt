# Pertemuan 4: Classification (Logistic Regression & Decision Tree)

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ============================================================
# Load dataset Iris
# ============================================================
# Dataset Iris digunakan untuk klasifikasi multi-kelas (3 spesies).
iris = load_iris()
# Split data: 80% training dan 20% testing.
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.2, random_state=42
)
```

```python
# Latih 2 model
# ============================================================
# 1. Logistic Regression: Menggunakan fungsi sigmoid untuk probabilitas.
lr = LogisticRegression(max_iter=200, multi_class='multinomial').fit(X_train, y_train)

# 2. Decision Tree: Membagi data berdasarkan kriteria seperti Gini Impurity.
dt = DecisionTreeClassifier(max_depth=3, random_state=42).fit(X_train, y_train)
```

```python
# Tugas 1 & 2: Evaluasi kedua model
# ============================================================
results = {}
for name, model in {"LogReg": lr, "DecisionTree": dt}.items():
    pred = model.predict(X_test)
    acc = round(accuracy_score(y_test, pred), 3)
    results[name] = acc
    
    print(f"\n{'='*50}")
    print(f"{name} — Accuracy: {acc}")
    print(f"{'='*50}")
```

```python
# Tugas 1 & 2: Evaluasi kedua model
# ============================================================
results = {}
for name, model in {"LogReg": lr, "DecisionTree": dt}.items():
    pred = model.predict(X_test)
    acc = round(accuracy_score(y_test, pred), 3)
    results[name] = acc
    
    print(f"\n{'='*50}")
    print(f"{name} — Accuracy: {acc}")
    print(f"{'='*50}")

    # Tugas 3: Confusion Matrix
    # Menampilkan tabel True Positive, True Negative, False Positive, dan False Negative.
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, pred))

    # Tugas 2: Precision, Recall, F1-Score
    # Precision: Ketepatan prediksi positif. Recall: Kemampuan mendeteksi kelas positif.
    print("\nClassification Report:")
    print(classification_report(y_test, pred, target_names=iris.target_names))
```

```python
# Tugas 4: Kesimpulan
# ============================================================
print("\n" + "=" * 50)
print("KESIMPULAN")
print("=" * 50)
best = max(results, key=results.get)
print(f"• LogReg accuracy       : {results['LogReg']}")
print(f"• DecisionTree accuracy : {results['DecisionTree']}")

if results["LogReg"] == results["DecisionTree"]:
    print(f"→ Kedua model memiliki accuracy yang sama ({results[best]}).")
    print("  Insight: Logistic Regression sangat efektif untuk dataset linear yang kecil,")
    print("  sementara Decision Tree memberikan interpretasi visual yang baik melalui struktur tree-nya.")
else:
    print(f"→ {best} memiliki accuracy lebih tinggi ({results[best]}).")

print("\n• Catatan: Decision Tree lebih rentan terhadap overfitting jika max_depth terlalu tinggi.")
print("• Catatan: Logistic Regression pada dataset ini menggunakan pendekatan multinomial untuk multi-kelas.")

print("\n✅ Pertemuan 04 selesai.")
```
