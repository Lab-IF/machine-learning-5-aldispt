# Pertemuan 6: Ensemble Methods (Random Forest & Gradient Boosting)

```python
import matplotlib
matplotlib.use("Agg")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer, load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, root_mean_squared_error, r2_score, classification_report

# ============================================================
# 1. Persiapan Data (Classification)
# ============================================================
data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)
```

```python
# Tugas 1: Perbandingan Single Model vs Ensemble (Bagging & Boosting)
# ============================================================
print("=" * 60)
print("Tugas 1: Perbandingan DT (Single) vs RF (Bagging) vs GB (Boosting)")
print("=" * 60)

#

models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42),
}

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"  {name:20s} Accuracy: {acc:.3f}")
```

```python
# Tugas 2: Feature Importance Analysis
# ============================================================
print("\n" + "=" * 60)
print("Tugas 2: Top 10 Feature Importance (Random Forest)")
print("=" * 60)

#

rf_model = models["Random Forest"]
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1][:10]

# Visualisasi Feature Importance
plt.figure(figsize=(10, 6))
plt.title("Top 10 Feature Importances - Random Forest")
plt.barh(range(10), importances[indices], align="center", color="skyblue")
plt.yticks(range(10), [data.feature_names[i] for i in indices])
plt.xlabel("Relative Importance")
plt.tight_layout()
plt.savefig("pertemuan06_feature_importance.png")
plt.close()

for rank, idx in enumerate(indices, 1):
    print(f"  {rank:2d}. {data.feature_names[idx]:30s}: {importances[idx]:.4f}")

print("\nVisualisasi disimpan ke: pertemuan06_feature_importance.png")
```

```python
# Tugas 3: Hyperparameter Tuning Analysis
# ============================================================
print("\n" + "=" * 60)
print("Tugas 3: Tuning Gradient Boosting (Learning Rate)")
print("=" * 60)

#

learning_rates = [0.001, 0.01, 0.1, 0.2]
for lr in learning_rates:
    gb_tuning = GradientBoostingClassifier(n_estimators=100, learning_rate=lr, random_state=42)
    gb_tuning.fit(X_train, y_train)
    acc = accuracy_score(y_test, gb_tuning.predict(X_test))
    print(f"  Learning Rate: {lr:<6} → Accuracy: {acc:.3f}")
```

```python
# Tugas 4: Ensemble Regression
# ============================================================
print("\n" + "=" * 60)
print("Tugas 4: Ensemble Regression (Random Forest Regressor)")
print("=" * 60)

db = load_diabetes()
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    db.data, db.target, test_size=0.2, random_state=42
)

rf_reg = RandomForestRegressor(n_estimators=100, random_state=42).fit(X_train_r, y_train_r)
y_pred_r = rf_reg.predict(X_test_r)

print(f"  RMSE: {root_mean_squared_error(y_test_r, y_pred_r):.3f}")
print(f"  R2 Score: {r2_score(y_test_r, y_pred_r):.3f}")
```

```python
# KESIMPULAN
# ============================================================
print("\n" + "=" * 60)
print("KESIMPULAN & INSIGHT")
print("=" * 60)
best_model = max(results, key=results.get)
print(f"1. Model terbaik adalah {best_model} dengan akurasi {results[best_model]:.3f}.")
print("2. Ensemble methods (RF & GB) terbukti meningkatkan performa dibandingkan")
print("   Single Decision Tree dengan cara mengurangi variance (RF) atau bias (GB).")
print(f"3. Fitur '{data.feature_names[indices[0]]}' adalah prediktor paling signifikan.")
print("4. Tuning Learning Rate pada Gradient Boosting sangat krusial; nilai yang")
print("   terlalu kecil (0.001) membuat model belajar terlalu lambat (underfitting).")

print("\n✅ Pertemuan 06 selesai.")
```
